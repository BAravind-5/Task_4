```python

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("/content/products_missing_demo (1).csv")
df.head()
```

```python
categorical_cols = df.select_dtypes(include='object').columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

categorical_cols, numerical_cols
```

Categorical Features: Product_Name, Category, Subcategory
Numerical Features: Price, Units_Sold, Revenue, Trend_Score

```python
le = LabelEncoder()
df['Category_Label'] = le.fit_transform(df['Category'])
df[['Category', 'Category_Label']].head()
```

Label Encoding is applied when categories have an implicit order
or when converting categories into numeric form for tree-based models.

```python
df_encoded = pd.get_dummies(
    df,
    columns=['Category', 'Subcategory'],
    drop_first=True
)

df_encoded.head()
```

One-Hot Encoding is used for nominal categorical features where
no natural order exists, preventing misleading ordinal relationships.

```python
scaler = StandardScaler()

num_cols = ['Price', 'Units_Sold', 'Revenue', 'Trend_Score']
df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])

df_encoded[num_cols].head()
```

```python
df[num_cols].describe()
```

```python
df_encoded[num_cols].describe()
```

Before scaling, numerical features had different ranges.
After scaling, features have mean  0 and standard deviation  1,
making the dataset suitable for distance-based algorithms.

Model Readiness:
- All categorical variables are encoded.
- Numerical features are scaled.
- Dataset is now suitable for algorithms such as
  Linear Regression, Logistic Regression, KNN, and SVM.

```python
df_encoded.to_csv("processed_dataset_task4.csv", index=False)
```
